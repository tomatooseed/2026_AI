```python
from typing import Any
def ask(user,system='You are a helpful assistant',max_tokens=100,temperature=0.2):
  #client = OpenAI(api_key=api_key)
  kwargs=dict[str,str | float | list[dict[str,str] | dict[str,Any]]](
    model='gpt-4o-mini',
    messages=[
      { 'role':'system','content':system},
      {'role':'user','content':user}
    ],
    temperature=temperature,
  )

  if max_tokens is not None : 
    kwargs['max_tokens']=max_tokens

  response=client.chat.completions.create(**kwargs)

  return response.choices[0].message.content

print(ask('서울 여행 코스 간단하게 1개만 추천해줘',max_tokens=100))
```

```python
from datetime import datetime

def get_city_time_basic():
  now=datetime.now().strftime('%Y-%m-%d %H:%M:%S')
  return now

print(get_city_time_basic())
```

```python
basic_tools=[
  {
    'type':'function',
    'function':{
      'name':'get_city_time_basic',
      'description': '현재 시간을 반환합니다'
       }
  }
]
BASIC_FUNCTIONS={
  'get_city_time_basic':get_city_time_basic,
}
```

```python
msg=[
    {
        'role': 'system',
        'content': 'You are a helpful assistant. Use tools when needed. 답변은 한국어로 해.',
    },
    {'role': 'user', 'content': '지금 서울이 몇 시인지 알려줘'},
]

response=client.chat.completions.create(
  model='gpt-4o-mini',
  messages=msg,
  tools=basic_tools,
  temperature=0.1,
)

print(response.choices[0].message)
temp=response.choices[0].message
print(temp.tool_calls)
```

```python

import json

msg.append(temp)


for tool_call in temp.tool_calls:
  fn_name=tool_call.function.name
  fn_args=json.loads(tool_call.function.arguments)

  print(f'호출 함수 : {fn_name}')
  print(f'인자:{fn_args}')


  result=BASIC_FUNCTIONS[fn_name](**fn_args)
  print(f'실행 결과 : {result}')


  msg.append(
    {'role':'tool',
    'tool_call_id': tool_call.id,
    'content': result
    }
  )


final=client.chat.completions.create(
  model ='gpt-4o-mini',
  messages = msg,
  temperature=0.1,
  tools=basic_tools
)

print(final.choices[0].message.content)
```

```python
#질문하고 tool찾고 답변까지 한방에 뱉어내는 function만들기

def run_agent_once(user_q,tools,tool_function,system_msg):
  messages=[
    {'role':'system','content':system_msg},
    {'role':'user','content':user_q}
  ]

  response=client.chat.completions.create(
    model='gpt-4o-mini',
    temperature=0.1,
    messages=messages,
    tools=tools
  )
  msg=response.choices[0].message

  if not msg.tool_calls:
    return msg.content

  messages.append(msg)

  for tool_call in msg.tool_calls:
    fn_name=tool_call.function.name
    fn_args=json.loads(tool_call.function.arguments)

    result=tool_function[fn_name](**fn_args)

    messages.append({
      'role':'tool',
      'tool_call_id':tool_call.id,
      'content': str(result) if result is not None else ''
    })

  final=client.chat.completions.create(
    model='gpt-4o-mini',
    messages=messages,
    temperature=0.1,
    tools=tools
  )

  return final.choices[0].message.content

```

```python
print(run_agent_once('뉴욕 현재 시간 알려줘',basic_tools,BASIC_FUNCTIONS,'You are a helpful assistant. 한국말로 대답하해. 내 질문에 맞는 tool을 call해서 활용해. '))
```

```python
#세계 시간 제대로 계산하는 tool 만들기 
import pytz

CITY_TZ = {
    '서울': 'Asia/Seoul',
    'seoul': 'Asia/Seoul',
    '뉴욕': 'America/New_York',
    'new york': 'America/New_York',
    '도쿄': 'Asia/Tokyo',
    'tokyo': 'Asia/Tokyo',
    '런던': 'Europe/London',
    'london': 'Europe/London',
}

def get_city_time_tz(city:str) -> str: 

  key=city.strip().lower()
  tz_name=CITY_TZ.get(key)
  if not tz_name:
    return json.dumps({'error':f'지원하지 않는 도시 : {city}'}, ensure_ascii=False)

  tz=pytz.timezone(tz_name)
  now=datetime.now(tz).strftime('%Y-%m-%d %H:%M:%S')
  return json.dumps({'city': city, 'time': now, 'timezone': tz_name}, ensure_ascii=False)

tz_tools=[
    {
      'type' : 'function',
      'function' : {
        'name' : 'get_city_time_tz',
        'description' : '도시의 시간대를 반영해 현재 시간을 반환합니다.',
        'parameters': {
          'type':'object',
          'properties':{'city':{'type':'string'}},
          'required':['city'],
        },
      },
    }
  ]


TZ_FUCTIONS={'get_city_time_tz':get_city_time_tz}
```

```python
basic_tools.extend(tz_tools)
BASIC_FUNCTIONS.update(TZ_FUCTIONS)
```

```python
import yfinance as yf

def get_stock_price(ticker:str) ->str:
  symbol=ticker.strip().upper()
  try:
    t=yf.Ticker(symbol)
    hist=t.history(period='2d')
    if hist.empty:
      return json.dumps({'error':f'{symbol} 데이터가 없습니다'},ensure_ascii=False)

    latest=hist.iloc[-1]
    prev_close=float(latest['Close']) if 'Close' in latest else None
    open_price=float(latest['Open']) if 'Open' in latest else None

    return json.dumps({
      'ticker':symbol,
      'open':round(open_price,2) if open_price is not None else None,
      'close':round(prev_close,2) if prev_close is not None else None,
      'currency':'USD',
      'source':'yfinance'
    },
    ensure_ascii=False,)
  except Exception as e:
    return json.dumps({'error':str(e)},ensure_ascii=False)
    
```

```python
yf_tools=[
    {
      'type' : 'function',
      'function' : {
        'name' : 'get_stock_price',
        'description' : '미국 주식 티커 심볼로 최근 시가,종가를 조회합니다.',
        'parameters': {
          'type':'object',
          'properties':{'ticker':{'type':'string'}},
          'required':['ticker'],
        },
      },
    }
  ]


YF_FUCTIONS={'get_stock_price':get_stock_price}

print('\n=== (3) yfinance 주식 Tool ===')
print(run_agent_once(
    '애플(AAPL)과 테슬라(TSLA) 최근 가격을 알려주고 간단 코멘트해줘.',
    yf_tools,
    YF_FUCTIONS,
    'Use tools when needed. 한국어로 간결하게 답변해.',
))
```

```python
basic_tools.extend(yf_tools)
BASIC_FUNCTIONS.update(YF_FUCTIONS)
```

```python
#여러개 tool중에서 내가 원하는 tool 가져올수 있게 만들기  

def run_agent_once2(user_q,tools,tool_function,system_msg):
  system_ad="""질문 내용에서 해당 사항이 포함된 경우 아래 tool을 call해서 답변해줘. 
  1. 특정 도시의 현재 시각을 물어보는 경우 ->  get_city_time_tz
  2. 주식 티커 혹은 주식 이름을 포함하여 주가를 물어보는 경우 -> get_city_time_basic
  해답을 모르는 경우 모른다고해"""
  messages=[
    {'role':'system','content':system_msg+system_ad},
    {'role':'user','content':user_q}
  ]

  response=client.chat.completions.create(
    model='gpt-4o-mini',
    temperature=0.1,
    messages=messages,
    tools=tools
  )
  msg=response.choices[0].message

  if not msg.tool_calls:
    return msg.content

  messages.append(msg)

  for tool_call in msg.tool_calls:
    fn_name=tool_call.function.name
    fn_args=json.loads(tool_call.function.arguments)
    print(f'호출 함수 : {fn_name}')

    result=tool_function[fn_name](**fn_args)

    messages.append({
      'role':'tool',
      'tool_call_id':tool_call.id,
      'content': str(result) if result is not None else ''
    })

  final=client.chat.completions.create(
    model='gpt-4o-mini',
    messages=messages,
    temperature=0.1,
    tools=tools
  )

  return final.choices[0].message.content
```
